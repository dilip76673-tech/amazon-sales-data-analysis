```python
sales_df = df.copy()
```

```python
sales_df = sales_df[sales_df["Status"] != "Cancelled"]
```

```python
sales_df = sales_df.drop(
    columns=["promotion-ids", "fulfilled-by"],
    errors="ignore"
)
```

```python
sales_df = sales_df.dropna(subset=["Amount"])
```

```python
sales_df = sales_df.dropna(
    subset=[
        "ship-city",
        "ship-state",
        "ship-postal-code",
        "ship-country"
    ]
)
```

```python
# Date column ko datetime me convert karna
sales_df["Date"] = pd.to_datetime(
    sales_df["Date"],
    format="%m-%d-%y"
)
```

```python
plt.figure(figsize=(10,6))

sales_df["Category"].value_counts().head(10).plot(
    kind="bar",
    color="skyblue"
)

plt.title("Top 10 Selling Categories")
plt.xlabel("Category")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(12,6))

sales_df["ship-state"].value_counts().head(10).plot(
    kind="bar",
    color="orange"
)

plt.title("Top 10 States by Orders")
plt.xlabel("State")
plt.ylabel("Number of Orders")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(12,6))

sales_df["ship-city"].value_counts().head(10).plot(
    kind="bar",
    color="green"
)

plt.title("Top 10 Cities by Orders")
plt.xlabel("City")
plt.ylabel("Number of Orders")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sales_df["Status"].value_counts().plot(
    kind="bar",
    color="red"
)

plt.title("Order Status Distribution")
plt.xlabel("Status")
plt.ylabel("Count")

plt.xticks(rotation=45)

plt.show()
```

```python
sales_df["Month Name"] = sales_df["Date"].dt.month_name()
```

```python
monthly_sales = sales_df.groupby("Month Name")["Amount"].sum()

monthly_sales = monthly_sales.reindex([
    "January","February","March","April",
    "May","June","July","August",
    "September","October","November","December"
])

plt.figure(figsize=(10,5))

monthly_sales.plot(marker="o")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")

plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

colors = sns.color_palette('pastel')[0:5]
plt.figure(figsize=(7, 7), dpi=100)

data = sales_df["Fulfilment"].value_counts()
explode = [0.05] + [0] * (len(data) - 1)

wedges, texts, autotexts = plt.pie(
    data, 
    labels=data.index, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=colors, 
    explode=explode, 
    shadow=True, 
    wedgeprops={'linewidth': 2, 'edgecolor': 'white'}, 
    textprops={'fontsize': 12, 'weight': 'bold'}
)

for autotext in autotexts:
    autotext.set_color('black')
    autotext.set_fontsize(11)

centre_circle = plt.Circle((0,0), 0.60, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Fulfilment Distribution", fontsize=16, weight='bold', pad=20)
plt.ylabel("")
plt.tight_layout()

plt.show()
```
